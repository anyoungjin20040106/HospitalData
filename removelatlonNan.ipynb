## 위도 경도 결측치 채우기

```python
import pandas as pd
from geopy.geocoders import Nominatim
geoloc = Nominatim(user_agent = '1234', timeout=None)    # 사용자 에이전트를 설정
```

```python
df=pd.read_excel('1.병원정보서비스 2024.12.xlsx')
df
```

```python
df=df[['요양기관명','종별코드명','시도코드명','시군구코드명', '읍면동','주소', '전화번호', '총의사수','좌표(X)', '좌표(Y)']]
df
```

```python
df.isnull().sum()
```

# 주소만 뽑는다
[출처](https://wonhwa.tistory.com/29)

```python
df['주소']=df['주소'].apply(lambda x :  " ".join(x.split(' ')[:4]))
df['주소']=df['주소'].apply(lambda x : x if x[-1]!=',' else x[:-1])
df
```

```python
import numpy as np

c = 0

def get_coordinates(address):
    global c
    try:
        location = geoloc.geocode(address)
        result = (location.longitude, location.latitude) if location else (np.nan, np.nan)
        c += 1
        print(f"{address} : {result} ({round(c/1488*100, 2)}%)")
        return result
    except Exception as e:
        print(f"Error for address {address}: {e}")
        return (np.nan, np.nan)

# 좌표가 NaN인 경우에만 지오코딩 실행
coords = df.apply(
    lambda row: get_coordinates(row['주소']) if pd.isna(row['좌표(X)']) and pd.isna(row['좌표(Y)']) else (row['좌표(X)'], row['좌표(Y)']),
    axis=1
)

df['좌표(X)'], df['좌표(Y)'] = zip(*coords)

df
```

```python
df.dropna(subset=['좌표(X)'],inplace=True)
df
```

```python
df.to_csv("removelatlonNan.csv",index=False)
```
